```python
import lsdb
import pandas as pd
from dask.distributed import Client

flares = lsdb.from_dataframe(pd.read_csv("flares_dr17.csv"))
tic = lsdb.read_hats(
    'https://data.lsdb.io/hats/tic/tic',
    margin_cache='https://data.lsdb.io/hats/tic/tic_10arcs',
    columns=["ID", "ra", "dec"],
)
```

```python
with Client() as client:
    df = flares.crossmatch(tic, radius_arcsec=0.5, suffixes=["", "_tic"]).compute()
df
```

```python
df.to_csv("flares_dr17_tic.csv")
```
